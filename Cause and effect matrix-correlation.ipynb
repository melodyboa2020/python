```python
import pandas as pd
import seaborn as sns
import numpy as np

cematrix = pd.read_csv("Cause and effect Matrix-correlation V2.csv")
cematrix = cematrix.drop(columns = ["No"], axis = 1)
cematrix.head()
```

```python
corr = cematrix.corr()
print(corr)

mask = np.triu(np.ones_like(cematrix.corr(), dtype=np.bool))
heatmap = sns.heatmap(cematrix.corr(),mask=mask, vmin=0, vmax=1, annot=True)
```
