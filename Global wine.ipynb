```python
from requests import get
url = 'https://www.sovos.com/shipcompliant/resources/wine-distribution-rules-by-state/'
response = get(url)
print(response.text[:1000])
```

```python
import bs4
from bs4 import BeautifulSoup
html_soup = BeautifulSoup(response.text, 'html.parser')
type(html_soup)
# bs4.BeautifulSoup
```

```python
state_containers = html_soup.find_all('div', class_ = 'three-column-content__content container')
print(type(state_containers))
print(len(state_containers))
```

```python
first_state = state_containers[0]
first_state
```

```python
firstname = first_state.p.text
firstname
```

```python
first_state.find_all('p')
```

```python
three_column = []
for i in first_state.find_all('p')[2:]:
    three_column.append(i.text)
print(three_column)
```

```python
states = []
three_columns = []

for container in state_containers:
    # state name
    state = container.p.text
    states.append(state)
    
    # three column content
    three_column = []
    for i in container.find_all('p')[2:]:
        three_column.append(i.text)
    three_columns.append(three_column)
```

```python
import pandas as pd
state_name = pd.DataFrame(states)
state_df = pd.DataFrame(three_columns)

state_df.iloc[14,0] = state_df.iloc[14,1]
state_df.iloc[14,1] = state_df.iloc[14,2]
state_df.iloc[14,2] = state_df.iloc[14,3]

state_df.iloc[15,0] = state_df.iloc[15,1]
state_df.iloc[15,1] = state_df.iloc[15,2]
state_df.iloc[15,2] = state_df.iloc[15,3]

state_df.iloc[33,0] = state_df.iloc[33,1]
state_df.iloc[33,1] = state_df.iloc[33,2]
state_df.iloc[33,2] = state_df.iloc[33,3]

state_df.iloc[35,0] = state_df.iloc[35,1]
state_df.iloc[35,1] = state_df.iloc[35,2]
state_df.iloc[35,2] = state_df.iloc[35,3]

state_df.iloc[30,0] = state_df.iloc[30,0].replace('Can I Self-Distribute to retailers?\n','') 
state_df.iloc[34,0] = state_df.iloc[34,0].replace('Can I Self-Distribute to retailers?\n','') 
state_df = state_df.drop([3],axis=1)
state_df.head()
```

```python
state_rule1 = pd.concat([state_name, state_df], axis=1)
state_rule1.columns = ['State', 'Is this a control state for wine?', 'Can I Self-Distribute to retailers?', 'Can I ship Direct to consumers?']
state_rule1.head()
```

```python
# scratch distribution process

distribution_containers = html_soup.find_all('div', class_ = 'two-column-content__content container container--small')
print(type(distribution_containers))
print(len(distribution_containers))
```

```python
first_distribution = distribution_containers[0]
first_distribution
```

```python
H4_1 = []
for c in distribution_containers:
    H4_1.append(c.find_all('div', class_= "two-column-content__col text-content")[2])
    
H4_1[0].find_all('p')
```

```python
two_column = []
two_column.append(first_distribution.find_all('p')[10].text)
two_column.append(first_distribution.find_all('p')[11].text)
two_column
```

```python
two_columns = []

for h in H4_1:
    two_column = []
    two_column.append(h.find_all('p'))
    two_columns.append(two_column)
    
two_columns[12][0][2]
```

```python
h1 = []
for i in range(51):
    h1.append(two_columns[i][0])
h1   
```

```python
H1 = []
for x in h1:

    test=[]
    hh = []
    for i in x:
        test.append(str(i))   

    for j in test:
        hh.append(j.strip('"<p>""</p>"'))
    H1.append(hh)
H1
```

```python
H1_df = pd.DataFrame(H1)
H1_df = H1_df.drop([0,3,4], axis=1)
H1_df.columns = ['DO DISTRIBUTORS NEED TO BE AWARE OF FRANCHISE RULES?', 'ARE THERE PRIMARY AMERICAN SOURCE RULES ON DISTRIBUTOR’S PURCHASES?']
H1_df.head()  
```

```python
wine_state_rule = pd.concat([state_rule1, H1_df], axis=1)
wine_state_rule.head()
```

```python
wine_state_rule.to_csv('state-rule.csv',index=False)
```
